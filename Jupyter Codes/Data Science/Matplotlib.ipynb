```python
x=np.linspace(0,10,10000)
y=np.sin(x)
fig=plt.figure()
ax=plt.axes()
```

```python
#Use pylab interface to draw
plt.plot(x,y)
plt.show()
```

```python
ax=plt.plot(x,y)
plt.show()
```

```python
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.plot(x,y)
ax2.plot(x,x**2-2*x-3)
```

```python
#Use numpy
x=np.linspace(0,10,1000)
plt.plot(x,np.cos(x))
```

```python
x=np.linspace(0,10,1000)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel('Truc X')
plt.ylabel('Truc y')
plt.title('Function sin(x)')
#plt.ylim(-0.5)
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='r') 
plt.plot(x,np.cos(x),color='k', linestyle='--')
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='r') 
plt.plot(x,np.cos(x),color='k', linestyle='--',alpha=0.1)
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='r',linestyle='-.',label='sin(x)') 
plt.plot(x,np.cos(x),color='k', linestyle='--',alpha=0.5, label='cos(x)')
plt.xlabel('SIN')
plt.ylabel('COS')
plt.legend()
plt.title('PHAT')
plt.show()
```
