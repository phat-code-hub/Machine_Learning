```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig=plt.figure()
ax=plt.axes()
```

```python
presidents_df=pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv',index_col="name")
presidents_df.head(n=10)
```

```python
plt.scatter(presidents_df['height'],presidents_df['age'])
plt.xlabel('H');
plt.ylabel('A')
plt.show()
```

```python
plt.scatter(presidents_df['height'],presidents_df['age'],
            marker='^',
            color='b')
plt.xlabel('height');
plt.ylabel('age')
plt.title('president')
plt.show()
```

```python
x=presidents_df['height']
y=presidents_df['age']
plt.xlabel('cao')
plt.ylabel('Tuoi')
plt.title('Quan he Tuoi- Cao')
plt.plot(x,y,'r^')
plt.show()
```

```python
#Use Pandas directly
presidents_df.plot(kind='scatter',
                   x='height',
                   y='age',
                   marker='^',
                   color='r',
                   figsize=(6,4),
                   title='U.S.Presidents')
plt.show()
```

```python
#Use Pandas directly
plt.style.use('ggplot')
presidents_df.plot(kind='scatter',
                   x='order',
                   y='height',
                   marker='^',
                   color='b',
                   figsize=(6,4),
                   title='U.S.Presidents')
plt.show()
```

##### HISTOGRAM

```python
presidents_df['height'].plot(kind='hist',
                             x='Cao',
                             y='Tan so',
                             bins=10,
                   title='U.S.Presidents\' Height')
plt.show()
```

```python
#The same result
plt.hist(presidents_df['height'],bins=10,color='b')
plt.xlabel('Tuoi')
plt.ylabel('Tan so')
plt.title('Count Height by person')
plt.show()
```

```python
presidents_df['age'].hist(bins=20)
```

```python
plt.hist(presidents_df['age'],bins=20)
```

##### BOXPLOT

```python
presidents_df.describe()
plt.style.use('ggplot')
presidents_df.boxplot(column='height');
print(presidents_df['height'].mean())
print(presidents_df['height'].median())
```

##### BARPLOT

```python
party_cnt=presidents_df['party'].value_counts()
plt.style.use('ggplot')
party_cnt.plot(kind='bar')
plt.show()
```

```python
party_cnt=presidents_df['age'].value_counts()
party_cnt.plot(kind='bar',color='b')
plt.show()
```

```python
party_cnt=presidents_df['age'].value_counts()
party_cnt.plot(kind='pie')
plt.show()
```
